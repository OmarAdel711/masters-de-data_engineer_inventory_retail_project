```sql
SELECT product_name,count(*) AS product_sold
FROM sales.order_items AS O
JOIN production.products AS P
ON O.product_id = P.product_id
GROUP BY product_name
ORDER BY product_sold DESC;
```

```sql
SELECT city, COUNT(*) AS CUST_CITY
FROM sales.customers AS C
JOIN sales.orders AS O
ON C.customer_id = O.customer_id
GROUP BY city
ORDER BY CUST_CITY DESC;
```

```sql
SELECT product_name,  count(*) AS product_needed
FROM sales.order_items AS OI
JOIN production.products AS P
ON OI.product_id = P.product_id
JOIN sales.orders AS O
ON OI.order_id = O.order_id
WHERE order_status IN (1, 2)
GROUP BY product_name
ORDER BY product_needed DESC;
```

```sql
SELECT product_name,  count(*) AS product_needed
FROM sales.order_items AS OI
JOIN production.products AS P
ON OI.product_id = P.product_id
JOIN sales.orders AS O
ON OI.order_id = O.order_id
WHERE order_status IN (1, 2)
GROUP BY product_name
ORDER BY product_needed DESC;
```

```sql
SELECT i.product_id,  count(*) AS product_needed
FROM sales.order_items AS OI
JOIN production.inventory_levels AS i
ON OI.product_id = i.product_id
JOIN sales.orders AS O
ON OI.order_id = O.order_id
WHERE order_status IN (1, 2)
GROUP BY i.product_id
ORDER BY product_needed DESC;
```
